# Задание лёгкого уровня сложности

Необходимо реализовать простейший вариант троянского коня.

**Внимание!!!**

Задание дано **только в образовательных целях**. Данные программы не стоит передавать кому-либо с целью нанесения вреда их системе.

## Про трояны

Троян - особый вид атаки на какую-либо систему. Данная атака подразумевает наличие какой-либо программы, которую хочет использовать пользователь, но помимо её на компьютер проникает какое-либо вредоносное ПО.

Они бывают различных видов:
1. Дающие контроль над компьютеорм жертвы
2. Уничтожающие данные
3. Создающие ботнет
4. Другие...

## Что необходимо реализовать

1. Какую-либо программу, которая будет "полезной" для пользователя: это может быть игра в крестики-нолики, калькулятор и т.д. *Желательно*, чтобы был какой-нибудь графический интерфейс.
2. Какой-либо вид троянской атаки по сети, например, дающее возможность отправить команды в консоль с удалённого сервера. Стоит быть осторожным, если вы будете реализовывать удаление файлов, в этом случае я **крайне рекомендую** использовать какой-либо вид песочницы, будь то виртуальная машина или контейнер.

## Описание хода работы

На мой взгляд, данную работу удобнее будет реализовывать в отдельных .py файлах. Поэтому вам даны файл для клиентской троянской программы (client.py), и сервера с которого будет идти работа с клиентом (server.py). Ниже представлена последовательность шагов для вида трояна с удалённым доступом.

1. Имеется запущенный удалённый сервер, который принимает клиентские соединения. 
2. Когда клиент запускает наше троянское приложение, у него в одном потоке запускается работа "полезной" программы, а в другом - создаётся соединение к нашему серверу. 
3. Затем с сервера приходят команды от "злоумышленника" и выполняются на клиентской машине.

## На что стоит обратить внимание

1. Стоит подумать над тем, как организовать работу сервера с клиентами.
2. Возможно, стоит реализовать не только команды в терминал, но и, например, просто вывод на консоль для упрощения первых этапов разработки. Если бедете реализовывать данный вариант, то можете подумать над тем, как хранить и проверять различные варианты работы команды.
