## Немного про эллиптические кривые и их примменение

В настоящее время они являются объектом исследований в области криптографии поскольку обеспечивают достаточно высокий уровень безопасности при небольшом количестве бит, которые мы тратим на ключи и саму кривую. Так же эллиптические кривые могут лежать в основе различных криптовалют (например, биткоин).

Эллиптической кривой называется кривая, задаваемая уравнением вида $y^2 + a_1xy + a_2y = x^3 + a_3x^2 + a_4x + a_6$. При этом, в криптографии используется другая форма, а именно форма Вейершстрасса: $y^2 = x^3 + Ax + B$. Мы будем работать именно с данной формой.

Стоит так же отметить, что мы рабтаем не над полем вещественных чисел, а над полем $\mathbb{Z}_p$, поэтому уравнение будет задаваться с небольшой модификацией: $y^2 = x^3 + Ax + B\; (mod p)$.

*Поясню*: эта запись означает, что мы сравниваем части равенсвта по модулю какого-то просого числа p.

Для криптографии необходимо знать сколько точек с координатами от (0, 0) до (p, p) то есть: (0, 0), (0, 1), (0, 2), ..., (0, p), (1, 0), (1, 1), (1, 2), ..., (1, p), ..., (p, 0), (p, 1), (p, 2), ..., (p, p) удоволеторяют нашему уравнению.
